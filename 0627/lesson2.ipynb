```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors = ['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values, #每個扇區的數值資料（list / array）
         labels=labels, #每個扇區的標籤文字
         colors=colors, #每個扇區的自訂顏色
         explode=[0.3, 0, 0, 0], #將第一個扇區「拉出」0.3 倍半徑距離，用來強調某個區塊
         shadow=True, #	在圓餅圖後方加上陰影，增加立體感
         autopct="%2.1f%%", #在扇區內部顯示百分比，格式為整數 2 位、小數 1 位（如 12.3%）
         startangle=180) #	第一個扇區的起始角度（度），180° 表示從正左方開始逆時針繪製（0° 為正右方）
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 若要顯示中文標籤，將字型設定為系統中文字型
font_path = r"C:\Windows\Fonts\msjh.ttc"
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['axes.unicode_minus'] = False

labels = ["奶茶", "紅茶", "綠茶", "咖啡"] #奶茶、紅茶、綠茶、咖啡
values = [50, 25, 35, 40] #50、25、35、40
colors = ['purple', 'pink', 'orange', 'darkblue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values, #每個扇區的數值資料（list / array）
         labels=labels, #每個扇區的標籤文字
         colors=colors, #每個扇區的自訂顏色
         explode=[0.2, 0, 0, 0], #將第一個扇區「拉出」0.2 倍半徑距離，用來強調某個區塊
         shadow=True, #	在圓餅圖後方加上陰影，增加立體感
         autopct="%1.1f%%", #在扇區內部顯示百分比，格式為整數 1 位、小數 1 位
         startangle=90) #	第一個扇區的起始角度（度），90° 表示從正左方開始逆時針繪製（0° 為正右方）
plt.show()
```
