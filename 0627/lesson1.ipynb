```python
import numpy as np
import math
import matplotlib.pyplot as plt # Matplotlib 是Python 中最熱門的2D 繪圖函式庫

t = np.arange(0, 2.5, 0.1)
#print(len(t))
t1 = t * math.pi #π
y1 = np.sin(t1)
y2 = np.cos(t1)
print(len(y1))

fig = plt.figure() #figure畫布
axes = fig.add_subplot(1,1,1) #1行1列一個圖表
axes.plot(t, y1) #畫圖 matplotlib.axes ，axes 表示「座標系統」
axes.plot(t, y2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 5, 0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi) 

figure = plt.figure()
ax1 = figure.add_subplot(2, 1, 1)
ax1.plot(t, y1, linestyle="--", color="red") #(t=X軸, y1=軸, linestyle="--"=樣式, color="red"=顏色)
ax2 = figure.add_subplot(2, 1, 2)
ax2.plot(t, y2, linestyle="-.",color="blue")
plt.show()
```

```python
# 步驟 1:匯入 numpy,並取別名 np

# 步驟 2:匯入 matplotlib.pyplot,並取別名 plt

# 步驟 3:用 np.arange() 產生 0 到 4(不含 4)、間隔 0.2 的陣列,存入 x

# 步驟 4:計算 y1 = cos(pi * x)(用 np.cos 與 np.pi)

# 步驟 5:計算 y2 = cos(2 * pi * x)

# 步驟 6:用 subplot 把版面切成 1 列 2 欄,選第 1 格(左圖)

# 步驟 7:畫出 x 與 y1,使用綠色實線

# 步驟 8:用 subplot 選第 2 格(右圖)

# 步驟 9:畫出 x 與 y2,使用洋紅色點線

# 步驟 10:顯示圖形
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 4, 0.2)
y1 = np.cos(t * np.pi)
y2 = np.cos(2 * t * np.pi)

figure = plt.figure()
ax1 = figure.add_subplot(1, 2, 1)
ax1.plot(t, y1, linestyle="-", color="green") #(t=X軸, y1=軸, linestyle="--"=樣式, color="red"=顏色)
ax2 = figure.add_subplot(1, 2, 2)
ax2.plot(t, y2, linestyle="-.",color="magenta")
plt.show()
```
