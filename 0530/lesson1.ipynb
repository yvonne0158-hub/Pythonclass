```python
#資料結構
#程式結構
#list
#list是可以修改
scores = [89,92,76,91,77]
print(scores)
print(scores[0]) #[]叫做subscript，0代表第一個數字89
print(scores[1])
print(scores[2])
print(scores[3])
print(scores[4])
scores[0] = 120
print(scores)

```

```python
#tuple
#tuple建立後就不能修改內容
#又叫read only 僅可閱讀
#何時使用,暫時儲存
scores1 = (85,92,97,73,59)
print(scores1)
print(scores1[0])
print(scores1[1])
print(scores1[2])
print(scores1[3])
print(scores1[4])
scores1[0] = 100
print(scores1)
```
