```python
# dictionary
students = {
	'chinese':89,
	'english':94,
	'math':65,
	'history':92,
	'discover': 95}
	
print(students)
print(students['chinese'])
print(students['math'])
print(students['discover'])
print(students['english'])
print(students['history'])
```

```python
scores1 = (85,92,97,73,59)
print(scores1)
print(scores1[0])
print(scores1[1])
print(scores1[2])
print(scores1[3])
print(scores1[4])
scores1[0] = 100
print(scores1)
```

```python
#list
#list,tuple串列資料
#for in 可以放串列資料、map資料，知道執行幾次才能用
scores1 = [89, 92, 76, 91, 77]
for s in scores1:
	print(s)
```

```python
#dictionary
students = {
	'chinese':89,
	'english':94,
	'math':65,
	'history':92,
	'discover': 95}
for key in students:
    print(key, students[key])
```
